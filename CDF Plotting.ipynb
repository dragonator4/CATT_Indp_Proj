```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from pylab import *
from PIL import Image
import os

#Edit the next ten lines to your liking.
main_path = 'C:/Projects/NPMRDS/' #This is the folder where NPMRDS data files are stored. Use UNIX style slashes
save_path = 'C:/Projects/NPMRDS/NPMRDS Figures/NRITS/' #This is the folder where the final graphs will be saved. Use UNIX style slashes
routes = ['NJ11-0019', 'VA08-0005', 'VA08-0014'] #Ensure that these are names of files. Should be a comma separated list. Ex: routes = ['VA08-0005', 'VA08-0012']
xy_limit = [10, 25, 10] #Axis limits for each corridor. Ensure that the number of limits is equal to the number of corridors. Must be a comma separated list. Ex: xy_limit = [10, 16]
marker = ['o', 'd', 's', '^', 'v', 'x', '*' 'H', 'd', 'h'] #The type of marker to rotate for the scatter plots. Must be a list. Ex: marker = ['o', 'd']
colors = ['blue', 'cyan' , 'magenta', 'blue', 'green', 'orange', 'purple'] #Add or remove, reorder to your preference
hr_hilt = [[16], [17], [7]] #The hours to highlight in the final CDF plots for each corridor. Enter hours from 00..23, in a nested list, with a nest for each corridor separated by commas. Ex: hr_hilt = [[7, 8], [11, 16, 17]]
columns = ['ttav', 'ttpc', 'ttft'] #Data column headers. Do not change.
vpp_len = [1.35, 1.74, 1.07]
npmrds_len = [1.3, 1.74, 1.07]

#Major and Minor ticks. Edit as desired.
#Scatter Plot Major ticks are at every hour
main_major_xticks = [datetime.time(x, 0, 0) for x in np.arange(0, 24, 1)]
main_major_xticks.append(datetime.time(23, 59))
#CDF major ticks are every 10%
cdf_major_xticks = np.arange(0, 1.01, 0.1)
#Major lables every 2 hours for main scatter plots
main_xlabels = ['00:00', '', '02:00', '', '04:00', '', '06:00', '', '08:00', '', '10:00', '', '12:00', '', '14:00', '', '16:00', '', '18:00', '', '20:00', '', '22:00', '', '23:59']
#Major lables every 3 hours for scatter plots accompanying cdf
cdf_xlabels = ['00:00', '', '', '03:00', '', '', '06:00', '', '', '09:00', '', '', '12:00', '', '', '15:00', '', '', '18:00', '', '', '21:00', '', '', '23:59']
#CDF minor ticks are every 5%
cdf_minor_xticks = np.arange(0, 1.01, 0.05)
#Minor ticks every 15 minutes for scatter plots
main_minor_xticks = []
for i  in range(0, 24):
    for j in np.arange(0, 60, 15):
        main_minor_xticks.append(datetime.time(hour=i, minute=j, second=0))

#Below this line, change with care. You have been warned.
cor_id = 0
for corridor in routes:
    pth = main_path + corridor + '.csv'
    df = pd.read_csv(pth, na_values=['\\N'])
    df['date_time'] = pd.to_datetime(df['date_time'])

    tmc = df['tmc'].unique()
    mdf = df[df['tmc'] == tmc[0]].copy()
    tmc = tmc[1:]
    for T in tmc:
        dft = df[df['tmc'] == T].copy()
        mdf = pd.merge(mdf, dft, on='date_time', how='inner', sort=False, copy=True)
        col_to_drop = list()
        for col in columns:
            mdf[col] = mdf[col + '_x'] + mdf[col + '_y']
            col_to_drop.append(col + '_x')
            col_to_drop.append(col + '_y')
        mdf.drop(col_to_drop, 1, inplace=True)
        
    #Normalize travel times
    for col in columns:
        mdf[col] = mdf[col] * (vpp_len[cor_id]/npmrds_len[cor_id])

    temp = pd.DatetimeIndex(mdf['date_time'])
    mdf['date'] = temp.date
    mdf['time'] = temp.time
    mdf['weekday'] = temp.weekday
    mdf = mdf[mdf['weekday'] < 5].copy()
    
    fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 5), dpi=600)
    ncolor = 0
    for col in columns:
        axes.plot(mdf['time'], (mdf[col]/60), marker[ncolor], color=colors[ncolor], label=col.upper())
        ncolor += 1
    axes.set_xlim(['00:00:00', '23:59:59'])
    axes.set_xticks(main_major_xticks)
    axes.set_xticks(main_minor_xticks, minor=True)
    axes.set_xticklabels(main_xlabels, fontsize=8, family='Times New Roman')
    axes.set_yticklabels(axes.get_yticks(), fontsize=8, family='Times New Roman')
    axes.legend(loc=0)
    axes.grid(b=True, which='major')
    axes.set_xlabel('Time of Day', fontsize=10, family='Times New Roman')
    axes.set_ylabel('Travel Time (Minutes)', fontsize=10, family='Times New Roman')
    axes.set_title('{0} - NPMRDS 24-hour Scatter Plot (Full Extent)'.format(corridor), fontsize=12, family='Times New Roman')
    figpath = save_path + corridor + '_Scatter_FullExt.png'
    fig.savefig(figpath, dpi=600, format='png')
    #Image.open(figpath).save((figpath[:-3] + 'jpg'),'JPEG')
    #os.remove(figpath)
    
    fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 5), dpi=600)
    ncolor = 0
    for col in columns:
        axes.plot(mdf['time'], (mdf[col]/60), marker[ncolor], color=colors[ncolor], label=col.upper())
        ncolor += 1
    axes.set_xlim(['00:00:00', '23:59:59'])
    axes.set_ylim([0, xy_limit[cor_id]])
    axes.set_xticks(main_major_xticks)
    axes.set_xticks(main_minor_xticks, minor=True)
    axes.set_xticklabels(main_xlabels, fontsize=8, family='Times New Roman')
    axes.set_yticklabels(axes.get_yticks(), fontsize=8, family='Times New Roman')
    axes.legend(loc=0)
    axes.grid(b=True, which='major')
    axes.set_xlabel('Time of Day', fontsize=10, family='Times New Roman')
    axes.set_ylabel('Travel Time (Minutes)', fontsize=10, family='Times New Roman')
    axes.set_title('{0} - NPMRDS 24-hour Scatter Plot (BT Data Extent)'.format(corridor), fontsize=12, family='Times New Roman')
    figpath = save_path + corridor + '_Scatter_Zoomed.png'
    fig.savefig(figpath, dpi=600, format='png')
    #Image.open(figpath).save((figpath[:-3] + 'jpg'),'JPEG')
    #os.remove(figpath)
    
    percentiles = np.arange(0.05, 1.00, 0.05)
    for jst in hr_hilt[cor_id]:
        for col in columns:
            fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5), dpi=600)
            for j in np.arange(0, 24, 1):
                sli = mdf[(mdf['time'] >= datetime.time(j, 0, 0, 0)) & (mdf['time'] <= datetime.time(j, 59, 59, 99))].copy()
                if len(sli[col].dropna()) > 0:
                    d = sli[col].describe(percentiles=percentiles)
                    if (j == jst):
                        axes[0].plot(sli['time'], (sli[col]/60), 'o', color='black')
                    else:
                        axes[0].plot(sli['time'], (sli[col]/60), 'o', color='light'+colors[0])
                    if (j == jst):
                        axes[1].plot((d[4:23]/60), percentiles, '-', color='black', linewidth=2)
                    else:
                        axes[1].plot((d[4:23]/60), percentiles, '-', color='light'+colors[0], linewidth=0.75)
            axes[0].set_xlim(['00:00:00', '23:59:59'])
            axes[0].set_ylim([0, xy_limit[cor_id]])
            axes[0].set_xticks(main_major_xticks)
            axes[0].set_xticks(main_minor_xticks, minor=True)
            axes[0].set_xticklabels(cdf_xlabels, fontsize=8, family='Times New Roman')
            axes[0].set_yticklabels(axes[0].get_yticks(), fontsize=8, family='Times New Roman')
            axes[0].grid(b=True, which='major')
            axes[0].set_xlabel('Time', fontsize=10, family='Times New Roman')
            axes[0].set_ylabel('Travel Time (Minutes)', fontsize=10, family='Times New Roman')
            axes[0].set_title('{0} - NPMRDS 24-hour Scatter Plot \nfor {1} at {2}:00 (BT Data Extent)'.format(corridor, col.upper(), jst), fontsize=12, family='Times New Roman')
            axes[1].set_ylim([0, 1])
            axes[1].set_xlim([0, xy_limit[cor_id]])
            axes[1].grid(True, which='major')
            axes[1].set_yticks(cdf_major_xticks)
            axes[1].set_yticks(cdf_minor_xticks, minor=True)
            axes[1].set_xticklabels(axes[1].get_xticks(), fontsize=8, family='Times New Roman')
            axes[1].set_yticklabels(axes[1].get_yticks(), fontsize=8, family='Times New Roman')
            axes[1].set_xlabel('Travel Time (Minutes)', fontsize=10, family='Times New Roman')
            axes[1].set_ylabel('Cumulative Probability (Percentile)', fontsize=10, family='Times New Roman')
            axes[1].set_title('{0} - NPMRDS Travel Time CDF Plot \nfor {1} at {2}:00 (BT Data Extent)'.format(corridor, col.upper(), jst), fontsize=12, family='Times New Roman')
            
            figpath = save_path + corridor + '_CDF_' + str(jst) + '_' + col.upper() + '.png'
            fig.savefig(figpath, dpi=600, format='png')
            Image.open(figpath).save((figpath[:-3] + 'jpg'),'JPEG')
            os.remove(figpath)
            
    cor_id += 1
```
